# Random Forest from `scikit-learn`

Imports

```python
# imports
import numpy as np
from sklearn import ensemble
from sklearn import preprocessing
```

Prepare Data

```python
x_raw = np.array([1, 2, 3])
y_raw = np.array([6, 4, 2])
x_scaler = preprocessing.MinMaxScaler((0, 1))
y_scaler = preprocessing.MinMaxScaler((0, 1))
x = x_scaler.fit_transform(x_raw.reshape(-1, 1))
y = y_scaler.fit_transform(y_raw.reshape(-1, 1))
```

Create Model Instance

```python
model = ensemble.RandomForestRegressor(
    n_estimators=3,
    bootstrap=True,
    verbose=2
)
```

Train

```python
model.fit(x, y)
```

Results

```python
# Predicting
x_predict = np.linspace(x_raw.min(), x_raw.max(), 10).reshape(-1, 1)
y_predict = y_scaler.inverse_transform(
    model.predict(
        x_scaler.transform(x_predict.reshape((-1, 1)))
    ).reshape(-1, 1)
)

# Plotting
import matplotlib.pyplot as plt

fig, ax = plt.subplots(constrained_layout=True)

ax.plot(x_raw, y_raw, ".")
ax.plot(x_predict, y_predict)
```
